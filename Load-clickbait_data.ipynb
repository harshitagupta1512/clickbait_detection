```python
## import json data from Data folder
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import jsonlines


# read the Training data
TrainData = []
with jsonlines.open('Data/clickbait17-train-170331/instances.jsonl') as reader:
    for obj in reader:
        TrainData.append(obj)
with jsonlines.open('Data/clickbait17-train-170630/instances.jsonl') as reader:
    for obj in reader:
        TrainData.append(obj)

TrainLabel = []
with jsonlines.open('Data/clickbait17-train-170331/truth.jsonl') as reader:
    for obj in reader:
        TrainLabel.append(obj)
with jsonlines.open('Data/clickbait17-train-170630/truth.jsonl') as reader:
    for obj in reader:
        TrainLabel.append(obj)
    

# read the Test data
TestData = []
TestLabel = []
with jsonlines.open('Data/clickbait17-test-170720/instances.jsonl') as reader:
    for obj in reader:
        TestData.append(obj)
with jsonlines.open('Data/clickbait17-test-170720/truth.jsonl') as reader:
    for obj in reader:
        TestLabel.append(obj)

```

```python
df = pd.DataFrame(TrainData)
for index, row in df.iterrows():
    appendedTargetParagraphs = "".join(row["targetParagraphs"]) 
    df.at[index, "appendedTargetParagraphs"] = appendedTargetParagraphs

dfTrain = df[["targetTitle", "appendedTargetParagraphs"]]

dfTrain.head()
```

```python
dfLabels = pd.DataFrame(TrainLabel)
dfLabels = dfLabels[["truthClass"]]
dfLabels.replace({"truthClass": {"clickbait": 1, "no-clickbait": 0}}, inplace=True)
dfLabels.head()

dfTrain = pd.concat([dfTrain, dfLabels], axis=1)
dfTrain.head()
```

```python
## convert data to csv
dfTrain.to_csv("Data/ClickBaitTrain.csv", index=False)
```

```python
df = pd.DataFrame(TestData)
for index, row in df.iterrows():
    appendedTargetParagraphs = "".join(row["targetParagraphs"]) 
    df.at[index, "appendedTargetParagraphs"] = appendedTargetParagraphs

dfTest = df[["targetTitle", "appendedTargetParagraphs"]]

dfTest.head()
```

```python
dfLabels = pd.DataFrame(TestLabel)
dfLabels = dfLabels[["truthClass"]]
dfLabels.replace({"truthClass": {"clickbait": 1, "no-clickbait": 0}}, inplace=True)
dfLabels.head()

dfTest = pd.concat([dfTest, dfLabels], axis=1)
dfTest.head()
```

```python
dfTest.to_csv("Data/ClickBaitTest.csv", index=False)
```
